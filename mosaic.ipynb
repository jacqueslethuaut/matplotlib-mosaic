```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
x = np.random.randn(1000)
y = np.random.randn(1000)

hist_data = np.random.randn(1000)

bar_labels = ['A', 'B', 'C', 'D']
bar_values = [10, 23, 17, 12]

pie_labels = ['Red', 'Blue', 'Green', 'Yellow']
pie_values = [15, 30, 45, 10]

line_x = np.linspace(0, 10, 100)
line_y = np.sin(line_x)

heatmap_data = np.random.randn(10, 10)

layout = [
    ['scatter', 'histogram'],
    ['bar', 'pie'],
    ['line', 'heatmap']
]

fig, axd = plt.subplot_mosaic(layout, figsize=(10, 10))

axd['scatter'].scatter(x, y, alpha=0.5)
axd['scatter'].set_title('Scatter plot')

axd['histogram'].hist(hist_data, bins=30, color='skyblue', edgecolor='black')
axd['histogram'].set_title('Histogram')

axd['bar'].bar(bar_labels, bar_values, color=sns.color_palette("pastel", 4))
axd['bar'].set_title('Bar chart')

axd['pie'].pie(pie_values, labels=pie_labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel", 4))
axd['pie'].set_title('Pie chart')

axd['line'].plot(line_x, line_y, label='sin(x)', color='r')
axd['line'].set_title('Line plot')

sns.heatmap(heatmap_data, ax=axd['heatmap'], cmap='YlGnBu')
axd['heatmap'].set_title('Heatmap')

plt.tight_layout()
plt.show()
```
